```python
# Buat 2 variable untuk menampung tulisan dalam satu paragraph (nilai 15)
# a. Doc1-19_001 berisi cerita kuliah yang disukai singkat dan bebas (misal “saya berkuliah di
# universitas trunojoyo jurusan teknik informatika. Selama saya menjalani kuliah ada matakuliah
# yang paling saya sukai karena …”). (lebih dari 75 kata)
# b. Doc2-19_001 berisi ambil dari 2-3 paragraf berita online terbaru. (lebih dari 75 kata)
# c. Gabungkan dua variable ini menjadi satu variable Doc3-19_001
# d. Tampilkan atau print ini dari dua variable tersebut
Doc1_19_001 = """Saya merupakan mahasiswa Teknik Informatika Angkatan 2019. Sejauh ini saya suka matakuliah Kecerdasan Komputasional, karena seru dalam pengimplementasiannya dan menurut saya lebih mudah diimplementasikan serta dipahami. Selain itu, saya memiliki kegiatan di lab setiap hari, hal ini membuat saya tidak bosan menjalani kuliah karena bertemu dengan banyak teman"""
Doc2_19_001 = """Acara perpisahan itu juga diramaikan dengan laga ekshibisi dengan format dua lawan dua, tiga lawan tiga, dan empat lawan empat. Greysia Polii berpartner dan bertanding melawan para bintang bulu tangkis dunia seperti Jongkolphan Kititharakul (Thailand), Misaki Matsumoto (Jepang), Shin Seung-chan (Korea Selatan), Thinaah Muralitharan (Malaysia), Huang Dong Ping (China), Ashwini Ponnappa (India), dan Gabriel Stoeva (Bulgaria). Selain nama-nama tersebut, hadir juga Sapsiree Tearattanachai (Thailand), Tai Tzu Ying (Taiwan), Yuta Watanabe (Jepang), Wang Yi-Lyu (China), Wang Chi-Lin (Taiwan), Lee Yang (Taiwan), Jia Yi Fan (China), dan Chen Qing Chen (China)"""

Doc3_19_001 = Doc1_19_001, Doc2_19_001
print(Doc3_19_001)
```

```python
# Lakukan proses tokenisasi kalimat terhadap variable gabungan yang telah dibuat, kemudian tampilkan atau print hasilnya (nilai 15)
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
  
word_tokens = word_tokenize(Doc3_19_001)
  
filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
filtered_sentence = []
  
for w in word_tokens:
    if w not in stop_words:
        filtered_sentence.append(w)
  
# print(word_tokens)
print("Hasil Tokenisasi :", filtered_sentence)
```

# **NO 3**

```python
# Lakukan proses menghapus angka, tanda baca dan stopword dari 2 variable yang telah dibuat, kemudian tampilkan atau print hasilnya. (nilai 20)
import nltk
from nltk.tokenize import word_tokenize


words1 = nltk.word_tokenize(Doc1_19_001)
words2 = nltk.word_tokenize(Doc2_19_001)

words1=[Word.lower() for Word in words1 if Word.isalpha()]
words2=[Word.lower() for Word in words2 if Word.isalpha()]

print("Hasil Doc1_19_001 : ", words1)
print("Hasil Doc2_19_001 : ", words2)
```

```python
# Lakukan proses untuk (nilai 25)
# a. Menghitung frekuensi kemunculan setiap tokens(kata) dalam teks.
# b. Menggambarkan frekuensi kemunculan setiap tokens dapat menggunakan library matplotlib
# c. Kemudian tampilkan atau print hasilnya.
import re
import string
print("Words1=Doc1_19_001 yang sudah di tokenisasi dan stopword")
frequency1 = {}
 
for word in words1:
    count = frequency1.get(word,0)
    frequency1[word] = count + 1
     
frequency1_list = frequency1.keys()
 
for words in frequency1_list:
    print (words, frequency1[words])

print("==================================================================================")
print("")
print("Words2=Doc2_19_001 yang sudah di tokenisasi dan stopword")
frequency2 = {}
 
for word in words2:
    count = frequency2.get(word,0)
    frequency2[word] = count + 1
     
frequency2_list = frequency2.keys()
 
for words in frequency2_list:
    print (words, frequency2[words])

import nltk
from nltk.corpus import stopwords
from nltk.classify import SklearnClassifier

from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
def wordcloud_draw(data, color = 'black'):
    words = ' '.join(data)
    cleaned_word = " ".join([word for word in words.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and not word.startswith('#')
                                and word != 'RT'
                            ])
    wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color=color,
                      width=2500,
                      height=2000
                     ).generate(cleaned_word)
    plt.figure(1,figsize=(13, 13))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
print ("HASIL")
print("words1")
wordcloud_draw(words1,'white')
print("")
print("words2")
wordcloud_draw(words2)
```
